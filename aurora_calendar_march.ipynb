```python
# Aurora Calendar Data Generator - March
import requests
from datetime import datetime
import json

# -------------------------
# 1️⃣ Fetch live NOAA Kp index
# -------------------------
url = "https://services.swpc.noaa.gov/json/planetary_k_index_1m.json"
try:
    data = requests.get(url).json()
    print("✅ NOAA Kp data fetched successfully.")
except Exception as e:
    print("❌ Failed to fetch NOAA data:", e)
    data = []

# -------------------------
# 2️⃣ Initialize calendar dictionary for March
# -------------------------
calendar_dict = {day: {} for day in range(1, 31)}

# -------------------------
# 3️⃣ Process Kp index for aurora likelihood
# -------------------------
for entry in data:
    try:
        dt = datetime.strptime(entry['time_tag'], '%Y-%m-%dT%H:%M:%S')
        if dt.month == 3:  # only March
            day = dt.day
            kp = float(entry['kp_index'])
            if kp >= 5:  # High aurora likelihood
                calendar_dict[day]['aurora'] = True
    except Exception as e:
        print(f"Skipping entry due to error: {e}")

# -------------------------
# 4️⃣ Pre-fill Blood Moon / Lunar Eclipse events
# -------------------------
# Example: Blood Moon on March 3
calendar_dict[3]['event'] = "Blood Moon"

# Example: Lunar Eclipse on March 20
calendar_dict[20]['event'] = "Lunar Eclipse"

# -------------------------
# 5️⃣ Pre-fill Meteor Showers / Lucky Stars
# -------------------------
meteor_days = [10, 15, 25]  # example dates
for day in meteor_days:
    calendar_dict[day]['meteor'] = True

# -------------------------
# 6️⃣ Export JSON for HTML calendar
# -------------------------
with open("aurora_calendar_march.json", "w") as f:
    json.dump(calendar_dict, f, indent=2)

from google.colab import files
files.download("aurora_calendar_march.json")

print("✅ JSON exported: aurora_calendar_march.json")
print("\nSample Output for March 1–15:")
print(json.dumps({k: calendar_dict[k] for k in range(1,11)}, indent=2))
```
